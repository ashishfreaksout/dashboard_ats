```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import numpy as np
import plotly.express as px

# Load data
df = pd.read_csv(r"/Users/ashishsharma/Documents/Python/processed_dataset.csv")
```

```python
# Scatter plot with Matplotlib
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Resume_Sentiment", y="Transcript_Sentiment", hue="decision")
plt.title("Resume vs Transcript Sentiment by Hiring Decision")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Resume_Sentiment'], bins=20, kde=True, color='blue', label="Resume Sentiment")
sns.histplot(df['Transcript_Sentiment'], bins=20, kde=True, color='red', label="Transcript Sentiment")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.title("Sentiment Analysis of Resumes & Transcripts")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Role', hue='decision', palette='viridis')
plt.xticks(rotation=45)
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.title("Job Roles vs Hiring Decisions")
plt.legend(title="Decision")
plt.show()

```

```python
text = " ".join(df['Lemmatized_Resume'].astype(str))
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Resumes")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='decision', y='Resume_Word_Count', hue='decision', palette='coolwarm', legend=False)
plt.xlabel("Hiring Decision")
plt.ylabel("Resume Word Count")
plt.title("Resume Word Count by Hiring Decision")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Resume_Sentiment', 'Transcript_Sentiment', 'Resume_Word_Count']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
